```
import pandas as pd
import numpy as np


#muat dataset titanic
URL="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
#untuk meihat tabel penumpang titanic (tambahan)
df.head()
```

```
#eksplorasi Awal
print(df.shape,df.columns.tolist())
print(df.isnull().sum())
```

```
# Filtering - penumpang kelas 1
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())
```

```
#groupby-tingkat keselamatan per kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean')
).round(3)
print(hasil)
```

Pertanyaan 1 : Berapa Jumlah baris dan kolom dataset titanic

Menggunakan perintah print(df.shape)

```
print(df.isnull().sum())
```

Dari hasil di atas diketahui bahwa kolom yang ada nilai kosong adalah Age, Cabin dan Embarked

Pertanyaan 3: Berapa persen penumpang yang selamat

Menggunakan persen_selamat = df['Survived'].mean() * 100


```
persen_selamat = df['Survived'].mean() * 100
print(persen_selamat,"%")
```

Dari hasil di atas diketahui 38% yang selamat

Pertanyaan 4 : Rata-rata usian penumpang wanita kelas 1 yang selamat



Menggunakan wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]

```
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Age'].mean())
```

Dari hasil di atas dapat diperoleh rata-rata usia wanita kelas 1 yang selamat adalah 34 tahun

Pertanyaan 5: Kelas peumpang mana yang tingkat keselamatan tertinggi

Menggunakan

```
#groupby-tingkat keselamatan per kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean')
).round(3)

print(hasil)
```

Dari hasil di atas diperoleh bahwa kelas 1 mempunyai tingkat keselamatan yang paling tinggi 0.63 atau 63%

Pertanyaan 6: Menampilkan hasil dari groupby('Pclass')['Survived'].mean()


```
hasil = df.groupby('Pclass')['Survived'].mean()
print(hasil)
```
